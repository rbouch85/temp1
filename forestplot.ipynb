# Forest Plot

```python
import pandas
import matplotlib.pyplot as plt
import plotnine as p9
import seaborn as sns
import numpy as np
```

```python
df = pandas.DataFrame({
    'label': [
        'Availability\nper 10-unit increase in ADI',
        'Accessibility\n(if available) \nper 10-unit increase in ADI'
    ],
    'estimate': [0.70, 0.94],
    'll': [0.56, 0.76],  # lower limit
    'hl': [0.84, 1.17]   # higher limit (upper limit)
    })

df
```

```python
# Create forest plot
(
    p9.ggplot(df, p9.aes(x='estimate', y='label'))
    + p9.geom_point(size=4, color='#2E86AB')
    + p9.geom_errorbarh(
        p9.aes(xmin='ll', xmax='hl'),
        height=0.25,
        color='#2E86AB',
        size=1
    )
    + p9.geom_vline(xintercept=1, linetype='dashed', color='#999999', alpha=0.5, size=0.8)
    + p9.labs(
        x='Odds Ratio',
        y=None
    )
    + p9.scale_x_continuous(limits=[0, 1.5], breaks=[0, 0.25, 0.5, 0.75, 1.0, 1.25, 1.5], expand=[0.02, 0.02])
    + p9.theme_minimal()
    + p9.theme(
        figure_size=(8, 5),
        axis_text_y=p9.element_text(size=16),
        axis_text_x=p9.element_text(size=14),
        axis_title_x=p9.element_text(size=14, margin={'t': 10}),
        axis_title_y=p9.element_blank(),
        panel_grid_major=p9.element_blank(),
        panel_grid_minor=p9.element_blank(),
        panel_border=p9.element_blank(),
        axis_line_x=p9.element_line(color='black', size=0.5)
    )
)
```
